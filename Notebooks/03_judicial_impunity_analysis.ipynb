Analizar el nivel global de impunidad judicial y entender qué factores se relacionan con casos no resueltos.

Preguntas

    - ¿Qué porcentaje de casos han sido resueltos?
    
    - ¿Qué porcentaje permanece impune?

    - ¿Qué regiones presentan mayor impunidad?

    - ¿Existe relación entre zonas de conflicto e impunidad?

Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargar dataset

```python
df = pd.read_csv(
    "../data/processed/cleaned_unesco.csv"
)

df.head()
```

Tasa global de resolución

```python

resolution_rate = (
    df["resolved"].mean() * 100
)

print(
    f"Tasa de resolución: {resolution_rate:.2f}%"
)
```

Tasa global de impunidad

```python

impunity_rate = (
    df["impunity"].mean() * 100
)

print(
    f"Tasa de impunidad: {impunity_rate:.2f}%"
)
```

Visualización resolución vs impunidad

```python
status = pd.Series({

    "Resueltos": df["resolved"].sum(),

    "Impunes": df["impunity"].sum()

})

status.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Comparación entre casos resueltos e impunes")

plt.ylabel("Número de casos")

plt.show()
```

Casos por categoría judicial

```python
df["case_category"].value_counts()
```

Impunidad por región

```python
region_impunity = df.groupby(
    "primary_region"
)["impunity"].mean() * 100

region_impunity.sort_values(
    ascending=False
)
```

Gráfico impunidad regional

```python
region_impunity.sort_values().plot(
    kind="barh",
    figsize=(10,6)
)

plt.title("Porcentaje de impunidad por región")

plt.xlabel("Porcentaje")

plt.show()
```

Zonas de conflicto

```python
pd.crosstab(

    df["Conflict Zone"],

    df["case_category"]

)
```

Tasa de resolución según conflicto

```python
conflict_resolution = df.groupby(
    "Conflict Zone"
)["resolved"].mean() * 100

conflict_resolution
```

Gráfico conflicto vs resolución

```python
conflict_resolution.plot(
    kind="bar"
)

plt.title(
    "Tasa de resolución según zona de conflicto"
)

plt.ylabel("Porcentaje")

plt.show()
```

Top países con mayor impunidad

```python
country_impunity = df.groupby(
    "Countries"
)["impunity"].mean() * 100

country_impunity.sort_values(
    ascending=False
).head(15)
```

# Hallazgos principales

Este análisis permite evaluar el desempeño judicial global frente a crímenes contra periodistas.

Principales hallazgos:

- La tasa global de impunidad continúa siendo extremadamente alta.
- Algunas regiones presentan tasas significativamente menores de resolución judicial.
- Los casos ocurridos en zonas de conflicto presentan menor probabilidad de resolución.
- Existen países donde prácticamente todos los casos permanecen sin resolver.

Conclusión:

La impunidad representa uno de los principales problemas estructurales en la protección internacional de periodistas.
