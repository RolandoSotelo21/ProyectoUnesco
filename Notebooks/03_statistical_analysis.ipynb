Import

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

Load

```python
df = pd.read_csv(
    "../data/processed/cleaned_unesco.csv"
)
```

Contingency table

```python
table = pd.crosstab(
    df["Conflict Zone"],
    df["resolved"]
)

table
```

Chi-square

```python
chi2, p, dof, expected = \
chi2_contingency(table)
```

Result

```python
print(p)
```

Interpretation

```python
if p < 0.05:

    print(
        "Conflict zone significantly affects resolution"
    )

else:

    print(
        "No statistical significance"
    )
```
