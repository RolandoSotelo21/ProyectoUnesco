Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

Load

```python
df = pd.read_csv(
    "../data/processed/cleaned_unesco.csv"
)
```

Cases Over Time

```python
cases = df.groupby("year").size()

plt.figure(figsize=(12,6))

plt.plot(
    cases.index,
    cases.values
)

plt.title(
    "Cases Over Time"
)

plt.show()
```

Top Countries

```python
top = (
    df["Countries"]
    .value_counts()
    .head(15)
)
```

```python
top.plot(kind="barh")

plt.title("Top Countries")

plt.show()
```

Gender

```python
df["Gender"].value_counts()
```

```python
df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.show()
```

Media Distribution

```python
df["Media"].value_counts().plot(
    kind="bar"
)

plt.title("Media Distribution")

plt.show()
```

Conflict Zone

```python
df["Conflict Zone"].value_counts().plot(
    kind="bar"
)

plt.title("Conflict Zone Cases")

plt.show()
```

Age Distribution

```python
plt.hist(
    df["Age"].dropna(),
    bins=20
)

plt.title("Age Distribution")

plt.show()
```

Map

```python
country_cases = (
    df.groupby("Countries")
      .size()
      .reset_index()
)

country_cases.columns = [
    "country",
    "cases"
]
```

```python
fig = px.choropleth(
    country_cases,
    locations="country",
    color="cases",
    locationmode="ISO-3"
)

fig.show()
```
