Evaluar el nivel de cooperación de los gobiernos frente a solicitudes oficiales realizadas por UNESCO.

Preguntas

- ¿Qué porcentaje de gobiernos responde a UNESCO?

- ¿Existe relación entre respuesta gubernamental y resolución judicial?

- ¿Qué regiones cooperan más?

Librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargar dataset

```python
df = pd.read_csv(
    "../data/processed/cleaned_unesco.csv"
)
```

Tasa de respuesta gubernamental

```python
response_rate = (
    df["government_responded"].mean() * 100
)

print(
    f"Tasa de respuesta gubernamental: {response_rate:.2f}%"
)
```

Casos con respuesta vs sin respuesta

```python
df["government_responded"].value_counts()
```

Visualización

```python
df["government_responded"].value_counts().plot(
    kind="bar"
)

plt.title(
    "Gobiernos que responden a UNESCO"
)

plt.ylabel("Número de casos")

plt.show()
```

Relación respuesta vs resolución

```python
pd.crosstab(

    df["government_responded"],

    df["case_category"]

)
```

Tasa de resolución según cooperación

```python
resolution_by_response = df.groupby(
    "government_responded"
)["resolved"].mean() * 100

resolution_by_response
```

Visualización

```python
resolution_by_response.plot(
    kind="bar"
)

plt.title(
    "Resolución judicial según cooperación gubernamental"
)

plt.ylabel("Porcentaje")

plt.show()
```

Cooperación por región

```python
regional_response = df.groupby(
    "primary_region"
)["government_responded"].mean() * 100

regional_response.sort_values(
    ascending=False
)
```

Gráfico regional

```python
regional_response.sort_values().plot(
    kind="barh",
    figsize=(10,6)
)

plt.title(
    "Cooperación gubernamental por región"
)

plt.xlabel("Porcentaje")

plt.show()
```

# Hallazgos principales

El nivel de cooperación institucional presenta diferencias importantes entre países y regiones.

Principales hallazgos:

- Muchos gobiernos no responden a solicitudes oficiales de UNESCO.
- La cooperación institucional parece asociarse a mejores tasas de resolución.
- Algunas regiones presentan niveles bajos de transparencia institucional.

Conclusión:

La falta de cooperación gubernamental puede dificultar procesos judiciales y aumentar niveles de impunidad.
