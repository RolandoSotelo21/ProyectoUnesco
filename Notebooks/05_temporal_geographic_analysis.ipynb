Analizar tendencias históricas y distribución geográfica global de asesinatos de periodistas.

Preguntas:

- ¿Los asesinatos aumentan o disminuyen con el tiempo?

- ¿Cuáles son los países más afectados?

- ¿Qué regiones concentran más casos históricos?

Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargar dataset

```python
df = pd.read_csv(
    "../data/processed/cleaned_unesco.csv"
)
```

Tendencia anual

```python
annual_cases = df.groupby(
    "year"
).size()

annual_cases
```

Evolución histórica

```python
annual_cases.plot(
    figsize=(12,6)
)

plt.title(
    "Evolución anual de asesinatos de periodistas"
)

plt.xlabel("Año")

plt.ylabel("Número de casos")

plt.show()
```

Distribución por década

```python
df["decade"].value_counts().sort_index()
```

Países con más casos

```python
top_countries = df["Countries"].value_counts().head(15)

top_countries
```

Top países gráfico

```python
top_countries.sort_values().plot(
    kind="barh",
    figsize=(10,6)
)

plt.title(
    "Países con mayor número de casos"
)

plt.xlabel("Número de casos")

plt.show()
```

Casos por región

```python
region_cases = df["primary_region"].value_counts()

region_cases
```

Región gráfico

```python
region_cases.plot(
    kind="bar",
    figsize=(10,5)
)

plt.title(
    "Distribución de casos por región"
)

plt.ylabel("Número de casos")

plt.show()
```

Hotspots globales

df["Countries"].value_counts().head(20)

Evolución por región a través del tiempo

```python
region_trend = pd.crosstab(

    df["year"],

    df["primary_region"]

)

region_trend.plot(
    figsize=(14,7)
)

plt.title(
    "Evolución histórica por región"
)

plt.xlabel("Año")

plt.ylabel("Número de casos")

plt.show()
```

# Hallazgos principales

El análisis temporal y geográfico permite identificar patrones estructurales globales.

Principales hallazgos:

- Existen períodos históricos con incrementos importantes en asesinatos.
- Algunos países concentran un volumen desproporcionadamente alto de casos.
- Ciertas regiones mantienen niveles persistentemente elevados de violencia contra periodistas.
- La distribución geográfica no es uniforme.

Conclusión:

La violencia contra periodistas presenta patrones históricos y regionales consistentes que requieren análisis permanente.
