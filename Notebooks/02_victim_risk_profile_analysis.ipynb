Analizar el perfil de las víctimas y entender qué factores laborales o personales se relacionan con mayor exposición al riesgo.

Preguntas

    - ¿Qué tipo de medio sufre más ataques?

    - ¿La mayoría de víctimas son periodistas locales?

    - ¿Existe diferencia entre staff y freelance?

    - ¿Cómo evoluciona la distribución por género?

Librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Cargar dataset limpio

```python
df = pd.read_csv(
    "../data/processed/cleaned_unesco.csv"
)

df.head()
```

Distribución por medio

```python

media_counts = df["Media"].value_counts()

media_counts
```

Gráfico por medio

```python
media_counts.plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Periodistas asesinados por tipo de medio")

plt.xlabel("Tipo de medio")

plt.ylabel("Número de casos")

plt.show()
```

Local vs extranjero

```python

df["Local"].value_counts()
```

Gráfico local vs extranjero

df["Local"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Periodistas locales vs extranjeros")

plt.ylabel("")

plt.show()

Staff vs freelance

```python

df["Staff"].value_counts(
    dropna=False
)
```

Staff según resolución del caso

```python
pd.crosstab(

    df["Staff"],

    df["case_category"]

)
```

Distribución por género

```python
df["Gender"].value_counts()
```

Evolución histórica por género

```python
gender_trend = pd.crosstab(

    df["year"],

    df["Gender"]

)

gender_trend.plot(
    figsize=(12,6)
)

plt.title(
    "Evolución histórica de víctimas por género"
)

plt.xlabel("Año")

plt.ylabel("Número de casos")

plt.show()
```

Cruce medio vs resolución

```python
pd.crosstab(

    df["Media"],

    df["case_category"]

)
```

# Hallazgos principales

Los resultados permiten identificar patrones importantes sobre el perfil de riesgo de las víctimas.

Principales hallazgos:

- La prensa escrita representa el grupo más afectado.
- Los periodistas locales concentran la mayoría de casos.
- Existen diferencias entre periodistas contratados y colaboradores independientes.
- La mayoría de víctimas históricamente corresponde al género masculino.
- Algunos sectores de medios muestran mayores niveles de impunidad que otros.

Conclusión:

La exposición al riesgo no afecta de manera uniforme a todos los perfiles periodísticos.
